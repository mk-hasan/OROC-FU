```python
def detect_outlier(data_1):
    outliers=[]
    #print(data_1)
    threshold=2
    mean_1 = np.mean(data_1)
    std_1 =np.std(data_1)  
    for y in data_1:
        z_score= (y - mean_1)/std_1 
        #print(np.abs(z_score))
        if np.abs(z_score) >= threshold:
            outliers.append(y)
    return outliers
```

```python
def same_number(df):
    #print(df)
    if(len(set(df))==1):
        return 1
    else:
        return 0
```

```python
def percentile_20(x):
    q1=x.quantile(0.2)
    q3=x.quantile(0.8)
    x =x.values.tolist()
    f_val=detect_outlier(x)
    return f_val
```

```python

data_outlier=data_t.groupby(by=["object1","object2"], as_index=False, sort=False)["similarity"].agg([percentile_20])

data_outlier_same= data_t.groupby(by="workerid", as_index=False, sort=False)["similarity"].agg([same_number])
```
